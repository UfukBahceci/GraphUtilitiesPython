# Problem 01

Support teams located in offices at certain fixed points in a city provide service throughout the day at pre-planned customer points that change daily according to demand. The company assigns support teams to these tasks every day. It is desired to reduce the total transportation costs of support teams to service points for each day. For each team, the Manhattan distance type seems appropriate for the distance between their offices and the customer locations they will serve. It has also been accepted to use distances for transportation costs. The following data is given for a specific day.

| Team no | Location(x,y) |
| :-----: | :-----------: |
| 0 | 12,18 |
| 1 | 15, 94 |
| 2 | 31, 35 |
| 3 | 34, 83 |
| 4 | 47, 61 |
| 5 | 53, 76 |
| 6 | 64, 20 |
| 7 | 77, 44 |
| 8 | 89, 63 |
| 9 | 94, 15 |


| Customer no | Location(x,y) |
| :---------: | :-----------: |
| 10 | 11, 45 |
| 11 | 54, 34 |
| 12 | 76, 29 |
| 13 | 91, 23 |
| 14 | 33, 78 |
| 15 | 65, 90 |
| 16 | 51, 76 |
| 17 | 42, 36 |
| 18 | 76, 16 |
| 19 | 24, 52 |

Now, we can define the positions(as the coordinates "x,y" on a 2D plane) of the nodes(vertices).

```python
vertices = np.array([[12, 18],
                     [15, 94],
                     [31, 35],
                     [34, 83],
                     [47, 61],
                     [53, 76],
                     [64, 20],
                     [77, 44],
                     [89, 63],
                     [94, 15],
                     [11, 45],
                     [54, 34],
                     [76, 29],
                     [91, 23],
                     [33, 78],
                     [65, 90],
                     [51, 76],
                     [42, 36],
                     [76, 16],
                     [24, 52]])
nodepos = vertices
```

```python
distancetype = 'Rectilinear'
drawnodes = True
writeedgelabels = False
```

```python
sizeoffirstvertexset = 10
G1 = Drawing.draw_bipartitesimplegraph(vertices, nodepos, distancetype, sizeoffirstvertexset, drawnodes, writeedgelabels)
```

To get information on the graph "G1":

```python
print(G1)
```

To get the sparse adjacency matrix (in compressed sparse row format) of the graph "G1":

```python
G1Adj = nx.adjacency_matrix(G1)
print(G1Adj)
```

To have the adjacency matrix of the graph "G1" in dense form:

```python
G1_dense = np.array(G1Adj.todense())
print(G1_dense)
```

```python
matching = bipartite.minimum_weight_full_matching(G1, top_nodes=None, weight='weight')
print(matching)
```
